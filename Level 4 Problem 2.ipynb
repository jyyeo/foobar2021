```python
import math
import copy
import numpy as np
def solution(dimensions, your_position, trainer_position, distance):
    #Your code here
    def get_pos(dimensions, your_position, trainer_position, distance):
        x_mirrors = int(math.ceil(float(your_position[0] + distance)/dimensions[0]))
        y_mirrors = int(math.ceil(float(your_position[1] + distance)/dimensions[1]))
#         print(x_mirrors, y_mirrors)
        
        # first quadrant
        your_positions, trainer_positions = [], []
        for i in range(x_mirrors):
            if i%2 == 1: # i is odd
                your_positions += [[(i+1)*dimensions[0]-your_position[0], your_position[1], 0]]
                trainer_positions += [[(i+1)*dimensions[0]-trainer_position[0], trainer_position[1], 1]]
            else:
                your_positions += [[i*dimensions[0]+your_position[0], your_position[1], 0]]
                trainer_positions += [[i*dimensions[0]+trainer_position[0], trainer_position[1], 1]]
        for j in range(y_mirrors):
            your_positions[j*x_mirrors:j*(x_mirrors+1)] = copy.deepcopy(your_positions[:x_mirrors])
            trainer_positions[j*x_mirrors:j*(x_mirrors+1)] = copy.deepcopy(trainer_positions[:x_mirrors])
        
        for j in range(y_mirrors):
            for i in range(x_mirrors):
                if j%2 == 1:
                    your_positions[j*x_mirrors+i][1] = (j+1)*dimensions[1]-your_position[1]
                    trainer_positions[j*x_mirrors+i][1] = (j+1)*dimensions[1]-trainer_position[1]
                else:
                    your_positions[j*x_mirrors+i][1] = j*dimensions[1]+your_position[1]
                    trainer_positions[j*x_mirrors+i][1] = j*dimensions[1]+trainer_position[1]
#         print(your_positions)
#         print(trainer_positions)
        positions = your_positions + trainer_positions
#         print(positions)
        
        # second quadrant
        tmp = copy.deepcopy(positions)
        tmp = [[coord[0], -coord[1], coord[2]] for coord in tmp]
        positions += tmp
#         print(positions)

        # third and fourth quadrants
        tmp = copy.deepcopy(positions)
        tmp = [[-coord[0], coord[1], coord[2]] for coord in tmp]
        positions += tmp
#         print(positions)
        return positions
    
    def calcDistance(your_position, curr_position):
        a, b = abs(your_position[0] - curr_position[0]), abs(your_position[1] - curr_position[1])
        return math.sqrt(a**2 + b**2)
    
    def calcAngle(your_position, curr_position):
        return np.arctan2(curr_position[1] - your_position[1], curr_position[0] - your_position[0])
    
    def filterDistance(your_position, positions, distance):
        filtered = {}
        for position in positions:
            dist, angle = calcDistance(your_position, position), calcAngle(your_position, position)
            if 0 < dist <= distance:
                if angle not in filtered or filtered[angle][1] > dist:
                    filtered[angle] = [position, dist]
        filtered_positions = [value[0] for key, value in filtered.items()]
#         print(filtered_positions)
        return filtered_positions
    
    def filterType(positions):
        trainers = [1 for position in positions if position[2] == 1]
        return len(trainers)

    positions = get_pos(dimensions, your_position, trainer_position, distance)
    filtered_by_dist = filterDistance(your_position, positions, distance)
    return filterType(filtered_by_dist)
```
